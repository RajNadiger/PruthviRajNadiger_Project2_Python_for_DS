```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy import stats
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data = pd.read_csv('renttherunway.csv')
```

```python
data = data.drop_duplicates()
```

```python
data
```

```python
data.describe()
```

```python
backupData = data
```

```python
data = backupData
```

```python
data = data.drop(['user_id', 'review_text'], axis=1)
```

```python
data['weight'] = data['weight'].str.replace('lbs', '').astype(float)

# data['rented_for'] = data['rented_for'].replace({'party: cocktail': 'party'})

data['height'] = data['height'].apply(lambda x: float(str(x).split(' ')[0].replace("'", ''))*12 + float(str(x).split(' ')[-1].replace('"', '')))

data.fillna(data.mean(), inplace=True)
data.fillna(data.mode().iloc[0], inplace=True)

numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
data = data[(np.abs(stats.zscore(data[numeric_cols])) < 3).all(axis=1)]
```

```python
data
```

```python
data['weight_height_ratio'] = data['weight'] / data['height']
data['review_month'] = pd.to_datetime(data['review_date']).dt.month

le = LabelEncoder()
categorical_features = data.select_dtypes(include=['object']).columns
for col in categorical_features:
    data[col] = le.fit_transform(data[col])

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data
```

```python
data.describe()
```

```python
pca = PCA(n_components=0.95)
data_pca = pca.fit_transform(data_scaled)

# K-means clustering
elbow_method = [KMeans(n_clusters=i, random_state=0).fit(data_pca).inertia_ for i in range(1, 11)]
plt.plot(range(1, 11), elbow_method)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

optimal_k = 5  # Assuming optimal K from the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=0).fit(data_pca)
print(f'Silhouette Score for K-means: {silhouette_score(data_pca, kmeans.labels_)}')
```

```python
data_pca.shape
```

```python
# Take a random sample of the data if it's too large
sample_size = 1000  # or whatever number your memory can handle
data_sample = []

for i in range(5):
    random_idx = np.random.choice(data_pca.shape[0], sample_size, replace=False)
    data_pca_sample = data_pca[random_idx]
    
    kmeans_sample = KMeans(n_clusters=optimal_k, random_state=0).fit(data_pca_sample)

    # Compute the dendrogram on the sampled data
    dendrogram(linkage(data_pca_sample, method='ward'))
    plt.title('Dendrogram')
    plt.xlabel('Samples')
    plt.ylabel('Euclidean distances')
    plt.show()

    agg_cluster = AgglomerativeClustering(n_clusters=optimal_k)
    agg_cluster.fit(data_pca_sample)
    print(f'Silhouette Score for Agglomerative Clustering: {silhouette_score(data_pca_sample, agg_cluster.labels_)}')

    data_sample = data.iloc[random_idx]


    data_sample['kmeans_cluster'] = kmeans_sample.labels_.tolist()
    data_sample['agg_cluster'] = agg_cluster.labels_.tolist()

    sns.pairplot(data_sample, vars=['weight_height_ratio', 'age', 'rating'], hue='kmeans_cluster')
    plt.show()
```
